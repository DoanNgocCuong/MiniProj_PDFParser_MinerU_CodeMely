### HOW TO RUN? 

1. Set up Env with python 3.10
Conda/.venv (search GPT for distinguish): In short, conda does more but is more complex. venv is simpler but only for Python.
```bash
conda create -n MinerU python=3.12
conda activate MinerU
```
or .venv (thường sẽ được tạo khi bạn pip install cái gì đó trong Cursor nó sẽ gợi ý tạo .venv. Tui thường xài cái này vì nó đơn giản và nhanh chóng trong Cursor)
```bash
python -m venv MinerU
source MinerU/bin/activate  # On macOS/Linux
MinerU\Scripts\activate     # On Windows

or simple
python -m venv .venv
.venv\Scripts\activate
```

```
pip install magic-pdf[full-cpu]

```


Lỗi này cho thấy `detectron2` chưa được cài đặt trong môi trường của bạn. `detectron2` là một thư viện phức tạp và không phải lúc nào cũng có sẵn trên PyPI để cài đặt trực tiếp bằng pip. Đây là một số bước bạn có thể thử để cài đặt `detectron2`:

1. Cài đặt từ nguồn được cung cấp:
   ```
   pip install detectron2 --extra-index-url https://myhloli.github.io/wheels/
   ```

2. Nếu bước trên không hoạt động, bạn có thể thử cài đặt từ GitHub repository của Detectron2:
   ```
   pip install 'git+https://github.com/facebookresearch/detectron2.git'
   ```

3. Nếu bạn gặp vấn đề với việc cài đặt từ GitHub, hãy thử cài đặt một phiên bản cụ thể:
   ```
   pip install detectron2==0.6
   ```

4. Nếu các phương pháp trên không hoạt động, bạn có thể cần phải cài đặt từ nguồn:
   ```
   git clone https://github.com/facebookresearch/detectron2.git
   cd detectron2
   pip install -e .
   ```

5. Đảm bảo rằng bạn đã cài đặt các dependencies cần thiết cho Detectron2:
   ```
   pip install torch torchvision opencv-python
   ```

6. Kiểm tra phiên bản Python của bạn. Detectron2 có thể yêu cầu một phiên bản Python cụ thể.

7. Nếu bạn đang sử dụng Windows, việc cài đặt Detectron2 có thể phức tạp hơn. Trong trường hợp này, bạn có thể cần phải sử dụng WSL (Windows Subsystem for Linux) hoặc một máy ảo Linux.

Sau khi thực hiện một trong các bước trên, hãy thử chạy lại lệnh `magic-pdf` của bạn. Nếu vẫn gặp vấn đề, hãy cung cấp thông báo lỗi mới để tôi có thể giúp bạn tiếp.
```

Tính tương thích: Các phiên bản khác nhau của torch và torchtext có thể có sự thay đổi trong cách mà chúng tương tác với nhau. Nếu hai phiên bản không tương thích, một số chức năng hoặc phương thức có thể không hoạt động đúng, dẫn đến lỗi như OSError: [WinError 127] The specified procedure could not be found.
2. Thay đổi trong API: Các phiên bản mới hơn có thể đã thay đổi API hoặc cách mà các thư viện tương tác với nhau. Nếu bạn đang sử dụng một phiên bản torchtext không tương thích với torch, có thể có các hàm hoặc lớp không còn tồn tại hoặc đã được thay đổi, gây ra lỗi khi cố gắng gọi chúng.
3. Yêu cầu phụ thuộc: Mỗi phiên bản của torch và torchtext có thể yêu cầu các thư viện phụ thuộc khác nhau. Nếu một thư viện phụ thuộc không được cài đặt hoặc không tương thích với phiên bản hiện tại, điều này có thể dẫn đến lỗi khi chạy mã.
4. Sửa lỗi trong phiên bản cũ: Đôi khi, các lỗi trong phiên bản mới có thể đã được sửa trong phiên bản cũ hơn. Nếu bạn gặp lỗi trong phiên bản mới, quay lại phiên bản cũ hơn có thể giúp bạn tránh được những vấn đề đó.

# ... các phần khác giữ nguyên ...

```bash
docker run your-image --path "/path/to/other.pdf" --output-dir "/other/output" --method "ocr"
```

Hoặc sử dụng biến môi trường trong Dockerfile:

```Dockerfile
# ... các phần khác giữ nguyên ...

ENV PDF_PATH="PDFTesting/testpdf2.pdf"
ENV OUTPUT_DIR="output_directory"
ENV METHOD="ocr"

ENTRYPOINT magic-pdf --path $PDF_PATH --output-dir $OUTPUT_DIR --method $METHOD
```

Sau đó, bạn có thể ghi đè các biến môi trường khi chạy container:

```bash
docker run -e PDF_PATH="/path/to/other.pdf" -e OUTPUT_DIR="/other/output" your-image
```

Những cách này sẽ giúp container của bạn linh hoạt hơn và có thể tái sử dụng cho nhiều file PDF khác nhau mà không cần build lại image.
```

Dựa trên yêu cầu của bạn về tính linh hoạt và khả năng tái sử dụng, tôi đề xuất sử dụng cách kết hợp ENTRYPOINT với biến môi trường. Đây là cách tôi khuyên bạn nên áp dụng:

1. Sửa đổi Dockerfile như sau:

````Dockerfile
FROM nvidia/cuda:11.8.0-runtime-ubuntu22.04

RUN apt-get update && apt-get install -y \
    python3.10 \
    python3-pip \
    && rm -rf /var/lib/apt/lists/*

RUN pip3 install --no-cache-dir --upgrade pip

WORKDIR /app

COPY requirements_fullCodeMely.txt .
RUN pip3 install --no-cache-dir -r requirements_fullCodeMely.txt
RUN pip3 install magic-pdf[full] --extra-index-url https://myhloli.github.io/wheels/

COPY . .

ENV CUDA_VISIBLE_DEVICES=0
ENV PDF_PATH="/app/PDFTesting/testpdf2.pdf"
ENV OUTPUT_DIR="/app/output_directory"
ENV METHOD="ocr"

ENTRYPOINT ["magic-pdf", "--path", "$PDF_PATH", "--output-dir", "$OUTPUT_DIR", "--method", "$METHOD"]
````

2. Tạo một script shell để chạy container:

Tạo một file `run_magic_pdf.sh` với nội dung sau:

````bash
#!/bin/bash

docker run --gpus all \
  -e PDF_PATH="$1" \
  -e OUTPUT_DIR="$2" \
  -e METHOD="$3" \
  -v "$(pwd)":/app \
  your-image-name
````

Giải thích:
- Sử dụng ENTRYPOINT thay vì CMD để đảm bảo `magic-pdf` luôn được chạy.
- Sử dụng biến môi trường để có thể dễ dàng thay đổi các tham số khi chạy container.
- Script shell giúp dễ dàng truyền các tham số khi chạy container.

Cách sử dụng:
1. Build Docker image:
   ```
   docker build -t magic-pdf-image .
   ```

2. Cấp quyền thực thi cho script:
   ```
   chmod +x run_magic_pdf.sh
   ```

3. Chạy container với các tham số tùy chỉnh:
   ```
   ./run_magic_pdf.sh "/path/to/your.pdf" "/path/to/output" "ocr"
   ```

Ưu điểm của cách này:
1. Linh hoạt: Dễ dàng thay đổi file PDF, thư mục output và phương thức xử lý.
2. Tái sử dụng: Có thể sử dụng cùng một image cho nhiều file PDF khác nhau.
3. Dễ sử dụng: Script shell giúp việc chạy container trở nên đơn giản.
4. An toàn: Sử dụng volume để chia sẻ dữ liệu giữa host và container, tránh mất dữ liệu.

Cách này cung cấp sự cân bằng tốt giữa tính linh hoạt và dễ sử dụng, phù hợp cho nhiều tình huống sử dụng khác nhau.
