```python
import fitz  # PyMuPDF
import os
import pandas as pd
import json
import re

# Hàm để trích xuất văn bản từ file PDF
def extract_text_from_pdf(pdf_path):
    pdf_document = fitz.open(pdf_path)
    text = ""
    for page_num in range(pdf_document.page_count):
        page = pdf_document.load_page(page_num)
        text += page.get_text()
    return text

# Hàm chia nhỏ văn bản thành các phần nhỏ hơn
def split_text(text, max_tokens=2000):
    words = text.split()
    current_chunk = []
    current_length = 0
    chunks = []

    for word in words:
        if current_length + len(word) + 1 > max_tokens:
            chunks.append(" ".join(current_chunk))
            current_chunk = [word]
            current_length = len(word) + 1
        else:
            current_chunk.append(word)
            current_length += len(word) + 1

    if current_chunk:
        chunks.append(" ".join(current_chunk))

    return chunks

# Hàm để làm sạch và tiền xử lý văn bản
def preprocess_text(text):
    # Xóa các URL
    text = re.sub(r'http\S+', '', text)
    
    # Tách các đoạn văn bản dựa trên các tiêu đề hoặc số thứ tự
    text = re.sub(r'(\d+\.\s)', r'\n\1', text)  # Thêm dòng mới trước các tiêu đề có số thứ tự
    text = re.sub(r'(II\.\s)', r'\n\1', text)  # Thêm dòng mới trước các tiêu đề "II."
    text = re.sub(r'(\n)', r'\n\n', text)  # Thêm dòng trống giữa các đoạn

    # Xóa các khoảng trắng thừa
    text = ' '.join(text.split())
    
    return text

# Hàm chính để trích xuất và xử lý văn bản từ PDF
def pdf_parser_with_gpt4(pdf_path):
    extracted_text = extract_text_from_pdf(pdf_path)
    preprocessed_text = preprocess_text(extracted_text)
    text_chunks = split_text(preprocessed_text)

    processed_chunks = []
    for chunk in text_chunks:
        processed_chunk = chunk  # Hiện tại, chỉ gắn đoạn văn bản vào processed_chunk mà không thay đổi gì
        processed_chunks.append(processed_chunk)

    return "\n".join(processed_chunks)

# Sử dụng hàm chính
file_path = r'D:\OneDrive - Hanoi University of Science and Technology\ITE10-DS&AI-HUST\Learn&Task\AI_Machine_Deep\AI-ML-DL\20232 NLP - M8 AIO\2024_PDF_Parser_CODEMELY\Tài liệu học kỹ thuật prompt của Group Bình dân học AI.pdf'
final_output = pdf_parser_with_gpt4(file_path)
print(final_output)

# Optional: Save the output to a JSON file
output_data = {
    "file_path": file_path,
    "processed_text": final_output
}

with open('output.json', 'w', encoding='utf-8') as json_file:
    json.dump(output_data, json_file, ensure_ascii=False, indent=4)

# Optional: Convert the output to a pandas DataFrame and save as CSV
df = pd.DataFrame([output_data])
df.to_csv('output.csv', index=False, encoding='utf-8')

# Optional: Save the output to a Markdown file
with open('output.md', 'w', encoding='utf-8') as md_file:
    md_file.write(f"# Processed Text from PDF\n\n")
    md_file.write(f"**File Path:** {file_path}\n\n")
    md_file.write(f"## Extracted and Processed Text\n\n")
    md_file.write(final_output)
```

```python
# Gắn LLMs vào lớp đằng sau

import fitz  # PyMuPDF
import openai
import os
import pandas as pd
import json

# Khởi tạo client OpenAI
client = openai.OpenAI(api_key=os.getenv("OPENAI_API_KEY"))

def get_completion(prompt, model="gpt-4o", temperature=0):
    messages = [{"role": "user", "content": prompt}]
    response = client.chat.completions.create(
        model=model,
        messages=messages,
        temperature=temperature,
    )
    return response.choices[0].message.content

# Hàm để trích xuất văn bản từ file PDF
def extract_text_from_pdf(pdf_path):
    pdf_document = fitz.open(pdf_path)
    text = ""
    for page_num in range(pdf_document.page_count):
        page = pdf_document.load_page(page_num)
        text += page.get_text()
    return text

# Hàm chia nhỏ văn bản thành các phần nhỏ hơn
def split_text(text, max_tokens=2000):
    words = text.split()
    current_chunk = []
    current_length = 0
    chunks = []

    for word in words:
        if current_length + len(word) + 1 > max_tokens:
            chunks.append(" ".join(current_chunk))
            current_chunk = [word]
            current_length = len(word) + 1
        else:
            current_chunk.append(word)
            current_length += len(word) + 1

    if current_chunk:
        chunks.append(" ".join(current_chunk))

    return chunks

# Hàm chính để trích xuất và xử lý văn bản từ PDF
def pdf_parser_with_gpt4(pdf_path):
    extracted_text = extract_text_from_pdf(pdf_path)
    text_chunks = split_text(extracted_text)

    processed_chunks = []
    for chunk in text_chunks:
        processed_chunk = get_completion(f"Extract important information from the following text:\n\n{chunk}")
        processed_chunks.append(processed_chunk)

    return "\n".join(processed_chunks)

# Sử dụng hàm chính
file_path = r'D:\OneDrive - Hanoi University of Science and Technology\ITE10-DS&AI-HUST\Learn&Task\PRODUCT_THECOACH\Bài 1\Tài liệu học kỹ thuật prompt của Group Bình dân học AI.pdf'
final_output = pdf_parser_with_gpt4(file_path)
print(final_output)

# Optional: Save the output to a JSON file
output_data = {
    "file_path": file_path,
    "processed_text": final_output
}

with open('output.json', 'w') as json_file:
    json.dump(output_data, json_file, ensure_ascii=False, indent=4)

# Optional: Convert the output to a pandas DataFrame and save as CSV
df = pd.DataFrame([output_data])
df.to_csv('output.csv', index=False, encoding='utf-8')
```
