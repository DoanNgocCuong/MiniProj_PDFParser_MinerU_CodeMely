```python
def export_chunks(output, chunks):
  with open(output, 'w') as f:
    for chunk in chunks:
      f.write(f"-----------page: {chunk.metadata['page']}------------------\n")
      f.write(f"content:{chunk.page_content}\n")
```

```python
import os, io
import fitz
from PIL import Image

def extract_images_from_pdf(pdf_path, lib='pymu'):
  # Output: list of image of each page[index]
    pdf = fitz.open(pdf_path)
    imgs = []
    for page in range(len(pdf)):
      page = pdf.load_page(page)
      image_list = page.get_images(full=True)
      lst = []
      for idx, img in enumerate(image_list):
        lst.append({})
        xref = img[0]
        base_image = pdf.extract_image(xref)
        image_bytes = base_image["image"]
        image_ext = base_image["ext"] if base_image["ext"] else "png"
        lst[idx]['bytes'] = image_bytes
        lst[idx]['ext'] = image_ext
      imgs.append(lst)
    return imgs
```

```python
import pytesseract
from io import BytesIO
import matplotlib.pyplot as plt
import numpy as np

def show_image(img):
  plt.imshow(np.asarray(img))
  plt.axis('off')  # Turn off axis numbers and ticks
  plt.show()

def OCR(image, lang, show=True):
    # If the image is a path or bytes, open it; otherwise, assume it's an Image object
    if isinstance(image, str) and image[0] == '/':
        img = Image.open(image)
    elif isinstance(image, bytes):
        img = Image.open(BytesIO(image))
    else:
        img = image

    if show:
        show_image(img)

    return pytesseract.image_to_string(img, lang=lang)
```

```python
def content_extract(chunks, imgs_pdf=None, show='False'):
  page_numbers = chunks[0].metadata['total_pages']
  i, j = 0, 0
  res, chunk_indx = [], []
  for idx in range(page_numbers):
    # print(idx, j, sep=' ')
    if j < len(chunks) and idx == chunks[j].metadata['page']:
      chunk_indx.append(chunks[j].page_content)
      j += 1
    else:
      chunk_indx.append('-1')
  # print(len(chunk_indx), chunk_indx, sep='\n')
  while(i < page_numbers):
    content = chunk_indx[i] + '\n' if chunk_indx[i] != '-1' else ''
    if imgs_pdf is not None:
      for idx in range(len(imgs_pdf[i])):
        content += OCR(imgs_pdf[i][idx]['bytes'], lang='vie', show=show) + '\n'
    res.append(content)
    i += 1
  return res
```

- PDF plumber có 2 phương thức cơ bản: extract_text(), extract_tables()
- Có thể lấy ảnh thông qua images -> xử lí ảnh lâu

```python
import pdfplumber

def extract_pdf_images_from_page(cur_page, show=True):
    images = cur_page.images
    img_lst = []
    for idx, image in enumerate(images):
        bbox = [
            image['x0'],
            cur_page.bbox[3] - image['y1'],
            image['x1'],
            cur_page.bbox[3] - image['y0']
        ]
        if all(0 <= coord <= cur_page.bbox[2 if i % 2 == 0 else 3] for i, coord in enumerate(bbox)):
            img_page = cur_page.crop(bbox=bbox)
            img_obj = img_page.to_image(resolution=500)
            img_lst.append(img_obj.original)
            if show:
                img_obj.show()
    return img_lst if len(img_lst) != 0 else None
def pdfPlumber_extract_images_from_pdf(pdf_path, show=True):
    page_imgs = []
    with pdfplumber.open(pdf_path) as pdf:
        for page_num, page in enumerate(pdf.pages):
            img_lst = extract_pdf_images_from_page(page, show=show)
            if img_lst is not None:
                page_imgs.append(img_lst)
            else:
                page_imgs.append('-1')
    return page_imgs
def pdfPlumber_texts_imgs(pdf_path):
    res = []
    imgs_lst= pdfPlumber_extract_images_from_pdf(pdf_path, show=False)
    img_txts = []
    for imgs in imgs_lst:
      page_imgs = ''
      for img in imgs:
        if img != '-1':
            page_imgs += OCR(img, lang='vie', show=False) + '\n'
      if len(page_imgs) != 0:
          img_txts.append(page_imgs)
      else:
          img_txts.append('-1')
    with pdfplumber.open(pdf_path) as pdf:
        for page_num, page in enumerate(pdf.pages):
            text_page = page.extract_text() + '\n'
            if img_txts[page_num] != '-1':
              text_page += img_txts[page_num] + '\n'
            else:
                res.append('')
            res.append(text_page)
    return imgs_lst, res

def pdfPlumber_extraction(pdf_path, extract='text'):
    with pdfplumber.open(pdf_path) as pdf:
        for page_num, page in enumerate(pdf.pages):
            if extract == 'text':
                print(page.extract_text())
            elif extract == 'table':
                print(page.extract_tables())
```
